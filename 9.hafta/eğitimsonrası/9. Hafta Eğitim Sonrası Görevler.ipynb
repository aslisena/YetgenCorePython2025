# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time
 
class Kumanda:
    def __init__(self, tv_durum = "Kapalı", ses_seviyesi=0, kanal_listesi=["CNBC"], kanal_ismi = "CNBC"):
        self.tv_durum = tv_durum
        self.ses_seviyesi = ses_seviyesi
        self.kanal_listesi = kanal_listesi
        self.kanal_ismi = kanal_ismi
    def tv_ac(self):
        if self.tv_durum == "Kapalı":
            print("Tv açılıyor...")
            self.tv_durum ="Açık"
        elif self.tv_durum == "Açık":
            print("Tv zaten açık")
            self.tv_durum= "Açık"

    def tv_kapat(self):
        if self.tv_durum == "Açık":
           print("Tv kapatılıyor...")
           self.tv_durum = "Kapalı"
        elif self.tv_durum == "Kapalı":
             print("Tv zaten kapalı")
             self.tv_durum = "Kapalı"
    
    def ses_ayarları(self):
        while True:
            print("Ses Seviyesi:", self.ses_seviyesi)
            ses_ayari = input("Ses seviyesini artırmak için +, azaltmak için - tuşlayın. Çıkmak için q tuşlayın:")
            if ses_ayari == "+":
                if self.ses_seviyesi < 25:
                    self.ses_seviyesi += 1
                    print("Ses seviyesi:", self.ses_seviyesi)
                else:
                    print("Ses seviyesi maksimumda")
            elif ses_ayari == "-":
                if self.ses_seviyesi > 0:
                    self.ses_seviyesi -= 1
                    print("Ses seviyesi:", self.ses_seviyesi)
                else:
                    print(" Ses seviyesi minimumda")
            elif ses_ayari == "q":
                print("Ses ayarları menüsünden çıkılıyor...")
                break 

    def kanal_ekle(self, kanal_ismi):
        time.sleep(1)
        if kanal_ismi in self.kanal_listesi:
            print(f"{kanal_ismi} kanalı zaten mevcut.")
            return
        self.kanal_listesi.append(kanal_ismi)
        print(f"{kanal_ismi} kanalı eklendi.")

    def kanal_sil(self, kanal_ismi):
        time.sleep(1)
        self.kanal_listesi.remove(kanal_ismi)
        print(f"{kanal_ismi} adlı kanal silindi.")

    def kanal_değiştir(self, kanal_ismi):
        time.sleep(1)
        if kanal_ismi in self.kanal_listesi:
            self.kanal_ismi = kanal_ismi
            print(f"{kanal_ismi} kanalına geçildi.")
        else:
            print(f"{kanal_ismi} kanalı mevcut değil.")

    def rastgele_kanal(self):
        time.sleep(1)
        rastgele_kanal = random.choice(self.kanal_listesi)
        self.kanal_ismi = rastgele_kanal
        print(f"{rastgele_kanal} kanalına geçildi.")

    def kanal_sayısı(self):
        print("Kanal sayısı:", len(self.kanal_listesi))
        print("Kanal listesi:", self.kanal_listesi)

    def __str__(self):
       return (f"TV Durumu:{self.tv_durum}, Ses Seviyesi: {self.ses_seviyesi}, Kanal listesi:{self.kanal_listesi},Şu anki kanal: {self.kanal_ismi}")

kumanda = Kumanda()

while True:
    print("\nİşlemler:")
    print("1. TV Aç")
    print("2. TV Kapat")
    print("3. Ses Ayarları")
    print("4. Kanal Ekle")
    print("5. Kanal Sil")
    print("6. Kanal Değiştir")
    print("7. Kanal Sayısını Görüntüle")
    print("8. Rastgele Kanal Seç")
    print("9. Kumanda Bilgileri")
    print("Çıkmak için 'q' tuşlayın.")
    
    islem = input("İşlemi Seçiniz: ")
    
    if islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarlari()
    elif islem == "4":
        kanal_isimleri = input("Eklemek istediğiniz kanal isimlerini ',' ile ayırarak giriniz: ")
        kanal_listesi = kanal_isimleri.split(",")
        for i in kanal_listesi:
            kumanda.kanal_ekle(i)
    elif islem == "5":
        print("Kanal sayısı:", len(kumanda.kanal_listesi))
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        print(kumanda)
    elif islem == "q":
        break
    else:
        print("Geçersiz işlem.")





               
```
